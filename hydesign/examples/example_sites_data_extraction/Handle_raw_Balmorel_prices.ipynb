```python
pr_df = pd.read_csv(f'{examples_filepath}example_sites_data_extraction/2030-EL_PRICE_raw.csv', parse_dates=True)
pr_df['Time'] = pd.DatetimeIndex(pr_df['Time'].apply(lambda x: x.replace('2030','2012')))
pr_df = pr_df.set_index('Time')

# Replace weather year index from 2030 to 2012
pr_df = pr_df.reindex(pd.date_range('2012-01-01 00:00','2012-12-31 23:00', freq='1h'),fill_value=np.NaN) 
```

```python
# Fill in missing dates with previous date
for t in list(pd.date_range('2012-02-29 00:00','2012-02-29 23:00', freq='1h')) + list(pd.date_range('2012-12-31 00:00','2012-12-31 23:00', freq='1h')):
    pr_df.loc[t,:] =  pr_df.loc[pd.to_datetime(t-pd.Timedelta('1d')),:]
    
pr_df.to_csv(f'{examples_filepath}Europe/2030-EL_PRICE.csv')
```
